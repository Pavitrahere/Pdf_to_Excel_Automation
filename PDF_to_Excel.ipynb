```python
import pdfplumber
import pandas as pd

# Path to your PDF file
pdf_path = r"C:\Users\USER\Downloads\PP_SSPA_INDUK_COMBINED.pdf"
# Output Excel path
output_excel = r"C:\Users\USER\Downloads\output.xlsx"

# Create an empty list to store all extracted tables
all_tables = []

# Open the PDF and extract tables from pages 1 to 91
with pdfplumber.open(pdf_path) as pdf:
    for i in range(0, 91):  # Index is 0-based, so 0 to 90 = pages 1 to 91
        page = pdf.pages[i]
        tables = page.extract_tables()
        for table in tables:
            if table:  # if not empty
                df = pd.DataFrame(table[1:], columns=table[0])  # First row as header
                all_tables.append(df)

# Combine all tables into one Excel file (one sheet)
with pd.ExcelWriter(output_excel, engine='openpyxl') as writer:
    for idx, df in enumerate(all_tables):
        sheet_name = f"Page_{idx+1}"
        df.to_excel(writer, index=False, sheet_name=sheet_name)

print("✅ PDF tables have been successfully exported to Excel.")
```
